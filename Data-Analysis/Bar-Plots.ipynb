### Author Bar Plot

```python
author_counts = getDataFrames.df_all['author'].value_counts()
author_counts
```

```python
active_author_counts = getDataFrames.df_save['author'].value_counts()
active_author_counts
```

```python
control_author_counts = getDataFrames.df_control['author'].value_counts()
control_author_counts
```

```python
custom_color = mcolors.CSS4_COLORS['darkseagreen']
custom_font = {'fontname': 'Georgia'}

top_n = 15  
top_author_counts = author_counts.nlargest(top_n)

plt.figure(figsize=(10, 6))
top_author_counts.plot(kind='bar', color=custom_color)
plt.title(f'Top {top_n} Authors by Frequency', **custom_font)
plt.xlabel('Author', **custom_font)
plt.ylabel('Frequency', **custom_font)
plt.xticks(rotation=45, ha='right', **custom_font)
plt.yticks(**custom_font)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({'control_counts': control_author_counts, 'active_counts': active_author_counts})

# Calculate the total counts for each author
df['total_counts'] = df['control_counts'] + df['active_counts']

# Sort the DataFrame based on total counts in descending order
df_sorted = df.sort_values(by='total_counts', ascending=False).head(15)

# Set font to Georgia
plt.rcParams['font.family'] = 'Georgia'

# Plot a stacked bar chart for the top 15 authors
plt.figure(figsize=(10, 6))
df_sorted[['control_counts', 'active_counts']].plot(kind='bar', stacked=True, color=['purple', 'darkseagreen'])

plt.title('Top 15 Authors - Stacked Bar Chart')
plt.xlabel('Authors')
plt.ylabel('Counts')
plt.legend(title='Type', bbox_to_anchor=(0.7, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

### Hashtags - all

```python
#drop NAs, split ",", transform to series, get value counts

hashtag_counts = getDataFrames.df_all['hashtag'].dropna().str.split(', ').explode().value_counts()

hashtag_counts
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(hashtag_counts)

colormap = 'plasma'

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Hashtags active

```python
#drop NAs, split ",", transform to series, get value counts

hashtag_active = getDataFrames.df_save['hashtag'].dropna().str.split(', ').explode().value_counts()

hashtag_active
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(hashtag_active)

colormap = 'plasma'

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
active_head = hashtag_active.head(15)
```

```python
import matplotlib.pyplot as plt
import numpy as np

hashtag_labels = active_head.index

# Set the bar height and position
bar_height = 0.35
bar_positions_active = np.arange(len(hashtag_labels))
bar_positions_control = bar_positions_active + bar_height

# Create the grouped bar chart
plt.figure(figsize=(6, 6))

plt.barh(bar_positions_active, active_head.values, height=bar_height, color='darkseagreen', label='Active Group')
plt.barh(bar_positions_control, control_head.values, height=bar_height, color='purple', label='Control Group')


plt.ylabel('Hashtags', fontname='Georgia')
plt.xlabel('Frequency', fontname='Georgia')
plt.title('Comparison of Top 15 Hashtag Frequencies between Active and Control Groups', fontname='Georgia')
plt.yticks(bar_positions_active + bar_height / 2, hashtag_labels, fontname='Georgia')
plt.legend()
plt.show()
```

## Summary Stats

```python
# Group by batch index (did not include in paper)
#grouped = df_all.groupby(["collectionTime", "testType"])

# Calculate summary statistics
summary_stats = getDataFrames.df_all.agg({
    "likes": ["mean", "median", "std", "min", "max"],
    "comments": ["mean", "median", "std", "min", "max"],
    "shares": ["mean", "median", "std", "min", "max"],
    "saves": ["mean", "median", "std", "min", "max"]
})

print("Summary Statistics for Numerical Variables:")
print(summary_stats)
```

```python
summary_stats.loc['mean'].plot(kind='bar', title='Average Values for Popularity Measures')
```

```python
mean_values = summary_stats.loc['mean']
median_values = summary_stats.loc['median']

mean_color = 'lightblue'
median_color = 'lightgreen'

bar_width = 0.35

mean_values.plot(kind='bar', color=mean_color, width=bar_width, position=0, label='Mean')

median_values.plot(kind='bar', color=median_color, width=bar_width, position=1, label='Median')

plt.title('Mean and Median Values for Popularity Measures')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
std_values = summary_stats.loc['std']

ax = std_values.plot(kind='bar', title='Standard Deviation Values of Variables of Interest', color='lightblue')


ax.get_yaxis().get_major_formatter().set_scientific(False)


plt.show()
```
