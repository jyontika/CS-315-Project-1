```python
# test data from group 2
df_test = pd.read_csv('02-08-19-22_like_by_random_data_liked_videos copy.csv')
df_test
```

```python
df_test1 = df_test[["batch", "index", "hashtag"]].copy()
df_test1
```

```python
# see what hashtags look like
print(list(df_test["hashtag"]))
```

```python
def process_hashtags(hashtag_str):
    if type(hashtag_str) == str:
        return hashtag_str.split(', ')
    else:
        return float("NaN")

#process_hashtags = lambda x: x.split(", ")

process_hashtags("fyp, bankrolldyl, funyyvideos, beautiful")
```

```python
df_test1["hashtag"] = df_test["hashtag"].apply(process_hashtags)
df_test1 = df_test1.explode("hashtag")
```

```python
df_test1
```

```python
# remove % from parsed hashtags (assume that anything after the % was not meant to be scraped)
filtered_percent = df_test1['hashtag'].notna() & df_test1['hashtag'].str.contains('%')
df_test1.loc[filtered_percent, 'hashtag'] = df_test1.loc[filtered_percent, 'hashtag'].str.split('%').str[0]


# separate out tagged accounts
filtered_at = df_test1['hashtag'].notna() & df_test1['hashtag'].str.contains('@')
df_tagged = df_test1[filtered_at]
df_test1.drop(df_test1.index[filtered_at], inplace=True)

df_test1
```

```python
df_tagged.rename(columns={'hashtag': 'tagged'}, inplace=True) 
# what do we want to do with this? how to handle nan?

df_tagged
```

```python
hashtags_list = list(df_test1["hashtag"])
print(hashtags_list)
```

```python
## recombine
df_test2 = (df_test1.groupby(['batch', 'index'])
      .agg({'hashtag': lambda x: x.tolist()}).reset_index())

df_test2
```

```python
batch_hashtags = df_test1["hashtag"].groupby(df_test1["batch"])
batch_hashtags.count()
```

```python
hashtag_list = list(df_test2['hashtag'])
hashtag_list
```

```python
from gensim.models import Word2Vec
```

```python
## try using word2vec skipgram model
# https://www.tensorflow.org/text/tutorials/word2vec

model = Word2Vec(sentences=hashtag_list, vector_size=100, window=5, min_count=1, sg=1)
```

```python
def get_similarity(hashtag1, hashtag2):
    try:
        similarity = model.wv.similarity(hashtag1, hashtag2)
        return similarity
    except KeyError:
        print("Error: either hashtag not found in model")
        return None 

# Example usage
hashtag1 = 'billieeilish'
hashtag2 = 'billie'
similarity_score = get_similarity(hashtag1, hashtag2)
print(f"Similarity between '{hashtag1}' and '{hashtag2}': {similarity_score}")
```
