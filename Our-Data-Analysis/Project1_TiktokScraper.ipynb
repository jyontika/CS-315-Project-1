```python
"""Given a username and password for a Tiktok account, open Tiktok email login, 
enter credentials, and manually perform security test
"""

def login(username, password):
    driver = webdriver.Firefox()
    driver.get('https://www.tiktok.com/login/phone-or-email/email')

    username_input = driver.find_element('xpath', '/html/body/div[1]/div/div[2]/div[1]/form/div[1]/input')
    password_input = driver.find_element('xpath','/html/body/div[1]/div/div[2]/div[1]/form/div[2]/div/input')

    username_input.send_keys(username)        
    password_input.send_keys(password)

    password_input.send_keys(Keys.RETURN)

    return driver

#driver = login('aytiktok315','tiktokCS315!')
#driver = login('compscirocks336@gmail.com','compscirocks336!')
#driver = login('johnplane876@gmail.com','iamjohn215!')
```

```python
def  findall_info(driver, batch_num):  
    all_details = []

    base_xpath = '/html/body/div[1]/div[2]/div[2]/div[1]'

    button_xpath = f'{base_xpath}//div[contains(@class, "css-1cifsuk-DivActionItemContainer")]/button'

    button_elements = driver.find_elements('xpath',button_xpath)

    num_buttons = 0
    video_details = []
    vid_num = 0
    for index, button_element in enumerate(button_elements, start=1):
        num_buttons+=1
        element = button_element.text
        if element =='Share':
            # Handle non-numeric case
            video_details.append(0) 

        if  num_buttons % 5 == 0:
            vid_num+=1
            all_details.append(video_details)
            username_path = f'/html/body/div[1]/div[2]/div[2]/div[1]/div[{vid_num}]/div/div[1]/div[1]/a[2]'
            element_user = driver.find_element('xpath',username_path)

            href = element_user.get_attribute('href')
            match = re.search(r'@(.*?)[\/\?]', href)
            if match:
                username = match.group(1)
                video_details.append(username)
            video_details.append(batch_num)
            video_details = []
            num_buttons = 0

        elif element !='Share':
            result = humanfriendly.parse_size(element)
            video_details.append(result)           
        #print(element)
    
    #print(f'Number of videos in this batch: {len(all_details)}')
    #print(all_details)
    return all_details
    


#index[0] = likes 
#index[1] = comments
#index[2] = saves
#index[3] = shares

all = findall_info(driver, 1)
print(all)
```

```python
def comdine_data(list1, list2):
    result = []

    for sublist1, sublist2 in zip(list1, list2):
        combined_list = [sublist1] + sublist2
        result.append(combined_list)

    #print(result)
    return result

#comdine_data(hash, all)
```

```python
""" Click one saved button on page
"""
def save_random(driver):
    base_xpath = '/html/body/div[1]/div[2]/div[2]/div[1]'

    button_xpath = f'{base_xpath}//div[contains(@class, "css-1cifsuk-DivActionItemContainer")]/button[3]'

    # Find all buttons with the specified button XPath
    button_elements = driver.find_elements('xpath',button_xpath)
    button_elements

    random_index = random.randint(0, len(button_elements) - 1)
    random_button = button_elements[random_index]
    random_button.click()

    vid_num = random_index+1
    username_path = f'/html/body/div[1]/div[2]/div[2]/div[1]/div[{vid_num}]/div/div[1]/div[1]/a[2]'
    element_user = driver.find_element('xpath',username_path)

    href = element_user.get_attribute('href')
    match = re.search(r'@(.*?)[\/\?]', href)
    if match:
        username = match.group(1)
        print(f'Username of the video we saved: {username}')
    return username
#save_random(driver)
```

```python
list_creds = [('compscirocks336@gmail.com','compscirocks336!'),(' compsciqueen4502@gmail.com',' BeOriginal6!'), ('user6252996274388', 'CS315Project!')]
#list_creds = [('tiktokcs315','tiktokCS315!'),('aytiktok315','tiktokCS315!')]

def main(credList, rounds, save):
    all_data = {}
    all_usersSaved = []
    username_saved = None

    random.shuffle(credList)
    for credential in credList:
        all_batches = []
        driver = login(credential[0], credential[1])
        time.sleep(25)
        count = 1
        checker = True
        while count <= rounds:
            print(f'========== Batch {count} ======')
            data_page = findall_info(driver, count)
            time.sleep(3)
            hashtags = get_hashtag(driver)
            time.sleep(3)
            finished_data = comdine_data(hashtags, data_page)
            time.sleep(3)
            all_batches.append(finished_data)
            print(f'Videos in Batch: {len(finished_data)}')
            if save and checker:
                user_saved = save_random(driver)
                time.sleep(3)
                username_saved =(credential[0],user_saved)
                all_usersSaved.append(username_saved)
                checker = False
            driver.refresh()
            time.sleep(5)
            count+=1
        
        all_data[credential[0]]= all_batches
        driver.close()

    return all_data, username_saved

save = True
all_data, user_saved = main(list_creds, 25, save)
print(all_data)
print(user_saved)
```

```python
def write(data):
    with open('output_data.csv', 'w', newline='') as csvfile:
        fieldnames = ['batch_user','batch_number','video_username','hashtags', 'likes', 'comments', 'saves','shares']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        # Write the header
        writer.writeheader()

        # Loop through the data dictionary
        for username, user_data in data.items():
            for data_entry in user_data:
                #print (data_entry)
                for entry in data_entry:
                    row_data = {
                        'batch_user': username,
                        'batch_number': entry[6],
                        'video_username': entry[5],
                        'hashtags': entry[0],  # Join hashtags into a string
                        'likes': entry[1],
                        'comments': entry[2],
                        'saves': entry[3],
                        'shares': entry[4],
                    }
                    writer.writerow(row_data)
write(all_data)
```
